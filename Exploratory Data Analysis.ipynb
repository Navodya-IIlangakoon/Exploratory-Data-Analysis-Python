```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data set
df=pd.read_csv("C:\\Users\\Hp\\OneDrive\\Desktop\\Codveda\\2) Stock Prices Data Set.csv")
```

```python
#Preview the dataset
print(df.head())
```

```python
#Convert date column to datetime
df['date']=pd.to_datetime(df['date'])
```

```python

#select numeric columns automatically
numeric_summary = df.describe(include=[float, int])
print(numeric_summary)
```

```python
# Median and mode
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]  # mode() returns DataFrame

print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)

# Standard deviation
std_dev = df.std(numeric_only=True)
print("\nStandard Deviation:\n", std_dev)
```

```python
#Histogram for Closing Prices
plt.figure(figsize=(8, 4))
sns.histplot(df['close'], bins=30, kde=True)
plt.title("Distribution of Closing Prices")
plt.xlabel("Close Price")
plt.ylabel("Frequency")

# Save plot as PNG
plt.savefig('histogram_close_price.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Box plot for volume
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['volume'])
plt.title("Boxplot of Volume")

# Save plot as PNG
plt.savefig('boxplot_volume.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Scatter plot for Close vs Volume
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='close', y='volume', alpha=0.6)
plt.title("Close Price vs Volume")
plt.xlabel("Close Price")
plt.ylabel("Volume")

# Save plot as PNG
plt.savefig('scatter_close_vs_volume.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Calculate correlation matrix
corr = df[['open', 'high', 'low', 'close', 'volume']].corr()

# Plot heatmap for correlations between numerical features
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")

# Save as PNG
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')

plt.show()
```
